```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from datetime import datetime
from numpy import nan
import string
```

```python
houses = pd.read_csv('Predicting-House-Prices-In-Bengaluru-Train-Data.csv')
```

```python
houses.head()
```

```python
houses.describe()
```

```python
pd.unique(houses['area_type'])
```

```python
houses['location'].nunique()
```

```python
area_type_dummies = pd.get_dummies(houses['area_type'],drop_first=True)
```

```python
houses = houses.join(area_type_dummies)
```

```python
houses
```

```python
houses.drop(columns='area_type',inplace=True)
```

```python
houses.head()
```

```python
#Reading the date from availability
availability_date = []
for item in houses['availability']:
    if(item[0].isdigit()):
        item = str(datetime.today().date().year) + '-' + item
        availability_date.append((datetime.strptime(item , '%Y-%d-%b')).date())
    else:
        availability_date.append(0)
```

```python
#Getting number of days from today till availability
days_till_available = []
for item in availability_date:
    if(item != 0):
        n_days = item - datetime.today().date()
        if(n_days.days > 0):
            days_till_available.append(n_days.days)
        else:
            days_till_available.append(n_days.days + 365)
    else:
        days_till_available.append(0)
```

```python
days_till_available[13318]
```

```python
houses['days_till_available'] = pd.Series(days_till_available , index=houses.index)
```

```python
houses.head()
```

```python
houses.drop(columns='availability',inplace=True)
```

```python
houses.head()
```

```python
houses.isnull().sum()
```

```python
#Removing columns with large number of missing values
houses.drop(columns='society',inplace=True)
```

```python
houses.head()
```

```python
size_bhk = []
size_bed = []
for i in range(len(houses['size'])):
    item = houses['size'][i]
    if(isinstance(item,str)):
        temp = item.split(' ')
        if('BHK' in temp[1]):
            size_bhk.append(int(item[0]))
            size_bed.append(0)
        elif('Bedroom' in temp[1]):
            size_bhk.append(0)
            size_bed.append(int(item[0]))
        elif('RK' in temp[1]):
            size_bhk.append(0)
            size_bed.append(0)
    else:
        size_bhk.append(0)
        size_bed.append(0)
```

```python
len(size_bhk)
```

```python
houses['BHK'] = pd.Series(size_bhk,index=houses.index)
houses['Bedroom'] = pd.Series(size_bed,index=houses.index)
```

```python
houses.head()
```

```python
houses.drop(columns='size',inplace=True)
#Dropping location column because highly correlated to area_type
houses.drop(columns='location',inplace=True)
```

```python
houses.head()
```

```python
houses.isnull().sum()
```

```python
#filling missing values
houses['bath'].fillna(np.mean(houses['bath']),inplace=True)
houses['balcony'].fillna(np.mean(houses['balcony']),inplace=True)
```

```python
houses.info()
```

```python
total_sqft_new = []
for item in houses['total_sqft']:
    if('-' in str(item)):
        temp = item.split()
        temp1 = (float(temp[0]) + float(temp[2]))/2
        total_sqft_new.append(temp1)
    elif(item.endswith("Sq. Meter")):
        temp = float(item[:-9])
        total_sqft_new.append(temp * 10.7639)    
    elif(item.endswith("Perch")):
        temp = float(item[:-5])
        total_sqft_new.append(temp * 272.25)
    elif(item.endswith("Acres")):
        temp = float(item[:-5])
        total_sqft_new.append(temp * 43560)
    elif(item.endswith("Sq. Yards")):
        temp = float(item[:-9])
        total_sqft_new.append(temp * 9)
    elif(item.endswith("Cents")):
        temp = float(item[:-5])
        total_sqft_new.append(temp * 435.6)
    elif(item.endswith("Guntha")):
        temp = float(item[:-6])
        total_sqft_new.append(temp * 1089)
    elif(item.endswith("Grounds")):
        temp = float(item[:-7])
        total_sqft_new.append(temp * 2400)
    else:
        total_sqft_new.append(float(item))
```

```python
houses['total_sqft_new'] = pd.Series(total_sqft_new, index=houses.index)
houses.drop(columns='total_sqft',inplace=True)
```
